# Visualizations

### Salary Distribution

We visualize the distribution of salaries to understand the concentration of earnings in tech roles.

```python
plt.figure(figsize=(8,5))
sns.histplot(df['salary_in_usd'], kde=True)
plt.title("Salary Distribution")
plt.xlabel("Salary (USD)")
plt.ylabel("Frequency")
plt.show()
```

### Company Size Distribution

Understand hiring dominance among small/medium/large companies.

```python
sns.countplot(df['company_size'])
plt.title("Company Size Distribution")
plt.ylabel("Count")
plt.show()
```

### Salary vs Experience Level

Shows how compensation increases with career seniority.

```python
group = df.groupby('experience_level')['salary_in_usd'].mean().reset_index()
sns.barplot(x='experience_level', y='salary_in_usd', data=group, palette='Set2', hue='experience_level')
plt.title("Average Salary by Experience Level")
plt.show()
```

### Salary Distribution by Experience Level (Violin Plot)

Shows how salary ranges expand and differ with career experience.

```python
plt.figure(figsize=(7,5))
sns.violinplot(x='experience_level', y='salary_in_usd', data=df, hue='experience_level', palette='husl')
plt.title("Salary Distribution by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Salary (USD)")
plt.show()
```

### Salary vs Remote Work Level (Scatter Plot)

Visualizing whether remote flexibility influences pay across experience levels.



```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='remote_ratio', y='salary_in_usd', hue='experience_level', data=df)
plt.title("Salary vs Remote Ratio by Experience Level")
plt.xlabel("Remote Ratio")
plt.ylabel("Salary (USD)")
plt.show()
```

### Top 5 Most Common Jobs (Countplot)

We keep only the most frequent roles to avoid overlapping text.

```python
top5 = df['job_title'].value_counts().head(5)

plt.figure(figsize=(7,4))
sns.countplot(y=df[df['job_title'].isin(top5.index)]['job_title'], palette='Set2')
plt.title("Top 5 Most Common Job Titles")
plt.xlabel("Count")
plt.ylabel("Job Title")
plt.show()
```

### Correlation Heatmap

Shows relationships between numerical variables.

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='twilight_shifted',linewidths=1.0, linecolor='lightgrey', fmt='.3f')
plt.title("Correlation Matrix")
plt.show()
```

### Salary Density Plot
Shows distribution smoothness over ranges.

```python
sns.kdeplot(df['salary_in_usd'])
plt.title("Salary Density")
plt.show()
```

### Experience Level Share in Dataset

```python
df['experience_level'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Experience Level Distribution")
plt.ylabel("")
plt.show()
```

### Pairplot: Salary & Remote Work Compared Across Top Job Roles

Relationships between key numerical features — salary and remote ratio — for the most common job titles only, to keep the chart clean and readable.


```python
# Select top 5 most common job titles to avoid clutter
top_roles = df['job_title'].value_counts().head(5).index
pair_df = df[df['job_title'].isin(top_roles)][['salary_in_usd', 'remote_ratio', 'job_title']]

# Create pairplot
sns.pairplot(pair_df, hue='job_title', diag_kind='kde')
plt.suptitle("Pairplot: Salary vs Remote Ratio by Top Job Titles", y=1.03)
plt.show()
```

### Salary Range per Experience Level (Box Plot)

Shows medians, quartiles, and outlier pay by experience.

```python
sns.boxplot(x='experience_level', y='salary_in_usd', data=df, hue='experience_level', palette='Greens')
plt.title("Salary Range by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Salary (USD)")
plt.show()
```

### Top 10 Highest Paying Job Titles

```python
top10_jobs = df.sort_values(by="salary_in_usd", ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x="salary_in_usd", y="job_title", data=top10_jobs, orient="h", hue='salary_in_usd', palette='Set3')
plt.title("Top 10 Highest Paying Job Titles (2024)")
plt.xlabel("Salary in USD")
plt.ylabel("Job Title")
plt.tight_layout()
plt.show()
```
